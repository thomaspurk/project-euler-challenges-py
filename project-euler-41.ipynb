```python
import sympy
import itertools
from tqdm.notebook import tqdm

all_digits = ['1','2','3','4','5','6','7','8','9']
answers = []

# Iterate over smaller and smaller digit sets
for l in tqdm(range(9,0,-1), desc='Iteration Set'):

    # Get the permutations of the current digit set
    # Each permutation is pandigital because the sub_list is pandigital
    sub_list = all_digits[:l]
    permutations = list(itertools.permutations(sub_list))

    # Iterate over the permutations and check each one for prime
    for p in tqdm(permutations, desc=f'Set Length: {l}'):
        p_int = int(''.join(p))
        if(sympy.isprime(p_int)):
            answers.append(p_int) # It's good, add it
            #print(f'\rPrime: {p_int:,}', end="")

print('')
print(f'Problem 41 Answer: {max(answers)}')
```
