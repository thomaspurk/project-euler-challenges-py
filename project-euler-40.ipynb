# Problem 40

<p>An irrational decimal fraction is created by concatenating the positive integers:
$$0.12345678910{\color{red}\mathbf 1}112131415161718192021\cdots$$</p>
<p>It can be seen that the $12$<sup>th</sup> digit of the fractional part is $1$.</p>
<p>If $d_n$ represents the $n$<sup>th</sup> digit of the fractional part, find the value of the following expression.
$$d_1 \times d_{10} \times d_{100} \times d_{1000} \times d_{10000} \times d_{100000} \times d_{1000000}$$</p>

```python
# This problem solutions simply builds the Champerone word through 1,000,000 characters in length
# There is no need for unit tests or functions
```

```python
n = 1
limit = 1_000_000
word = "0" # Start with zero due to python's zero based indexing
# Champernowne's word goes to infinity
# We only need enough characters to satisfy the queries in the question
while len(word) <= limit:
    word += str(n)
    n+=1

answer = int(word[1]) * int(word[10]) * int(word[100]) * int(word[1_000]) * int(word[10_000]) * int(word[100_000]) * int(word[1_000_000])

print(f'Problem 40 Answer: {answer}')
```
