# Problem 33

<p>$145$ is a curious number, as $1! + 4! + 5! = 1 + 24 + 120 = 145$.</p>
<p>Find the sum of all numbers which are equal to the sum of the factorial of their digits.</p>
<p class="smaller">Note: As $1! = 1$ and $2! = 2$ are not sums they are not included.</p>

### From ChatGPT

> Definition of "Strong" Numbers: A "strong number" is a number for which the sum of the factorials of its digits equals the number itself. Because factorials grow very fast, there are only finitely many strong numbers.
Upper Bound Reasoning: Since 9! = 362880, for a number with d digits, the maximum sum of the factorials of digits would be ùëë √ó 362880. To have a chance of being a strong number, this sum must at least be equal to the number itself.

```python
import math

n = 10

# Track the possible numbers
numbers = []

# Given the number of digits . . .
#   - what is the maximum sum assuming all digits are 9
#   - and the smallest number that can be made from those digits
num_digits = len(str(n))
max_factorial_sum = num_digits * math.factorial(9)
smallest_number = 10**(num_digits-1)

# if max_factorial_sum < smallest_number
# Then the number is too big, further cases are not possible

while max_factorial_sum >= smallest_number:
    if( n % 10000 == 0):
        print(f'\r {n}, Count: {len(str(n))}', end='')
    int_list = list(map(int,str(n)))
    fact_list = list(map(math.factorial, int_list))
    
    fact_sum = sum(fact_list)
    
    if(fact_sum == n):
        numbers.append(n)

    # Prepare for next loop
    n += 1
    num_digits = len(str(n))
    max_factorial_sum = num_digits * math.factorial(9)
    smallest_number = 10**(num_digits-1)

print('')
print(f'Problem 34 Answer: {sum(numbers)}')    
```
