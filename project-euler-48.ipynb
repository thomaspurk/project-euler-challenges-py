# Problem 48 - Self Powers

<p>The series, $1^1 + 2^2 + 3^3 + \cdots + 10^{10} = 10405071317$.</p>
<p>Find the last ten digits of the series, $1^1 + 2^2 + 3^3 + \cdots + 1000^{1000}$.</p>


```python
# This solution is simple and does not require functions or unit test
```

```python
from tqdm.notebook import tqdm
limit = 1000
value = 0
for n in tqdm(range(1, limit + 1)):
    value += n**n

print(f'Problem 48 Answer: {str(value)[-10:]}')
```
